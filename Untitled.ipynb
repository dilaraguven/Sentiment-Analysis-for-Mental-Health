```python
df.drop('Unnamed: 0',axis=1,inplace =True)
df.head()
```

```python
sum_of_null_values = df.isnull().sum()
sum_of_null_values
```

```python
df_cleaned = df.dropna()
df_cleaned
```

```python
df_cleaned.loc[:, 'statement'] = df_cleaned['statement'].str.lower()
```

```python
df_cleaned.loc[:, 'statement'] = df_cleaned['statement'].str.replace(r'[^a-zA-Z0-9\s]', '', regex=True)
df_cleaned
```

```python
df_cleaned.isnull().sum()
```

```python
import seaborn as sns

# Count plot
plt.figure(figsize=(10, 6))
sns.countplot(x='status', data=df_cleaned, palette='pastel', hue='status', legend=False)
plt.title('Status Dağılımı - Count Plot')
plt.xlabel('Status')
plt.ylabel('Sayı')
plt.show()
```

```python
df_cleaned.describe()
```

```python
# En çok geçen 10 statement değerini bul
top_10_statements = df_cleaned['statement'].value_counts().head(10)
top_10_statements_df = top_10_statements.reset_index()
top_10_statements_df.columns = ['statement', 'count']
top_10_statements_df
```

```python
# 'statement' değeri "oh my gosh" olan satırları filtrele
filtered_df = df_cleaned[df_cleaned['statement'] == 'niki best boy']
filtered_df
```

```python
df_cleaned = df_cleaned.copy()  # DataFrame'in bir kopyasını oluştur
df_cleaned.loc[:, 'statement_length'] = df_cleaned['statement'].apply(lambda x: len(x.split()))
df_cleaned
```

```python
df_cleaned.loc[:, 'char_length'] = df_cleaned['statement'].apply(lambda x: len(x))
df_cleaned
```

```python
# Kelime çeşitliliği (unique kelime sayısı / toplam kelime sayısı)
df_cleaned['word_diversity'] = df_cleaned['statement'].apply(lambda x: len(set(x.split())) / len(x.split()) if len(x.split()) > 0 else 0)
df_cleaned
```

```python
from textblob import TextBlob

# Duygu skoru ekleme (TextBlob ile)
df_cleaned['sentiment'] = df_cleaned['statement'].apply(lambda x: TextBlob(x).sentiment.polarity)
df_cleaned
```

```python
def get_sentiment_label(score):
    if score > 0:
        return 'positive'
    elif score < 0:
        return 'negative'
    else:
        return 'neutral'

df_cleaned['sentiment_label'] = df_cleaned['sentiment'].apply(get_sentiment_label)
df_cleaned
```

```python
# Status'e göre statement uzunluğu ortalaması
plt.figure(figsize=(10, 6))
sns.barplot(x='status', y='statement_length', data=df_cleaned, palette='coolwarm', hue='status', legend=False)
plt.title('Status\'e Göre Ortalama Statement Uzunluğu')
plt.xlabel('Status')
plt.ylabel('Ortalama Kelime Sayısı')
plt.show()
```

```python
# Status'e göre kelime çeşitliliği violin plot'u
plt.figure(figsize=(10, 6))
sns.violinplot(x='status', y='word_diversity', data=df_cleaned, palette='muted', hue='status', legend=False)
plt.title('Status\'e Göre Kelime Çeşitliliği Dağılımı')
plt.xlabel('Status')
plt.ylabel('Kelime Çeşitliliği')
plt.show()
```

```python
# Status ve sentiment_label arasındaki ilişkiyi gösteren bar plot
plt.figure(figsize=(10, 6))
sns.countplot(x='status', hue='sentiment_label', data=df_cleaned, palette='viridis')
plt.title('Status ve Sentiment Label Dağılımı')
plt.xlabel('Status')
plt.ylabel('Sayı')
plt.show()
```

```python
from sklearn.preprocessing import LabelEncoder

# LabelEncoder'ı oluştur
le = LabelEncoder()

# 'status' ve 'sentiment_label' sütunlarını sayısal hale getir
df_cleaned['status_encoded'] = le.fit_transform(df_cleaned['status'])
df_cleaned['sentiment_label_encoded'] = le.fit_transform(df_cleaned['sentiment_label'])
df_cleaned
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# TF-IDF ile metni sayısal verilere dönüştürme
vectorizer = TfidfVectorizer(max_features=5000)
X_tfidf = vectorizer.fit_transform(df_cleaned['statement'])
```

```python
from scipy.sparse import hstack

# Diğer sayısal özellikleri numpy array formatına dönüştürme
X_numerical = df_cleaned[['statement_length', 'char_length', 'word_diversity', 'sentiment', 'sentiment_label_encoded']].values

# TF-IDF ve sayısal özellikleri birleştirme
X_combined = hstack([X_tfidf, X_numerical])
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report

# Veriyi eğitim ve test setlerine ayırma
X_train, X_test, y_train, y_test = train_test_split(X_combined, df_cleaned['status_encoded'], test_size=0.2, random_state=42)

# Modeli oluştur ve eğit
model = RandomForestClassifier()
model.fit(X_train, y_train)

# Test setinde tahmin yap
y_pred = model.predict(X_test)

# Modelin performansını değerlendir
print("Accuracy:", accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```python
from xgboost import XGBClassifier

model = XGBClassifier()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)

# Performans değerlendirme
print("Accuracy:", accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```python
from sklearn.metrics import confusion_matrix

# Karışıklık matrisi oluşturma
cm = confusion_matrix(y_test, y_pred)

# Karışıklık matrisini görselleştirme
plt.figure(figsize=(10, 7))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=model.classes_, yticklabels=model.classes_)
plt.xlabel('Predicted')
plt.ylabel('True')
plt.title('Confusion Matrix')
plt.show()
```
